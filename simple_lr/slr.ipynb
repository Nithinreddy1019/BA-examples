```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("Hitters1.csv")

print(df.head())
```

```python
#Mark the explanatory and dependent variable
x=df.iloc[:, 0].values
y=df.iloc[:, 1].values

x,y
```

```python
#Check whether there is a linear relation between x & y

plt.scatter(x,y)
```

```python
#reshape data using array.reshape(-1,1) - if you have one ffeature

regressor = LinearRegression()
regressor.fit(x.reshape(-1, 1), y)

```

```python
print(regressor.coef_)
print(regressor.intercept_)
```

```python
y_pred = regressor.predict(x.reshape(-1, 1))
print(y_pred)

r2_score(y, y_pred)
```

```python
y_new = np.sqrt(y)
print(y_new)
```

```python
regressor2 = LinearRegression()
regressor2.fit(x.reshape(-1, 1), y_new)
```

```python
print(regressor2.coef_)
print(regressor2.intercept_)

y_pred2 = regressor2.predict(x.reshape(-1, 1))
print(y_pred2)

r2_score(y_new, y_pred2)
```
